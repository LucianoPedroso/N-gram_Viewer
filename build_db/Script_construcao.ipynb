```python
# Função para ler os textos de cada tese de mestrado
def open_text(arq):
    text = []
    aux = []
    a = open(arq, encoding='utf8')
    linhas = a.readlines()
    for linha in linhas:
        text.append(linha)
    return text
```

## Limpeza dos Dados

```python
# Função para limpar os \n dos textos
def limpeza_dados_001(texto):
    aux = []
    aux_final = []
    for i in range(len(texto)):
        if texto[i] != '\n':
            a = texto[i]
            aux.append(a)
    return aux
```

```python
text_trato_001 = []
for i in range(len(textos)):
    a = limpeza_dados_001(textos[i])
    b = [string.replace('\n', '').replace('\x0c', '').replace('.', '').replace("\ufeff", '').replace(" \t", '') for string in a] 
    text_trato_001.append(b)
```

## Identificando Localização nos textos

```python
texto = []
categorias = []
for i in range(len(df['Texto'])):
    texto_aux = []
    categorias_aux = []
    
    qtd = len(df['Texto'][i])
    print(i)
    
    if qtd >= 1000000 :
        tamanho = len(df['Texto'][i]) // 3  # tamanho de cada pedaço
        
        text_parte1 = str(df['Texto'][i][:tamanho])  # primeira parte
        text_parte2 = str(df['Texto'][i][tamanho:tamanho*2])  # segunda parte
        text_parte3 = str(df['Texto'][i][tamanho*2:])  # terceira parte
        
        doc_parte1 = nlp(text_parte1)
        doc_parte2 = nlp(text_parte2)
        doc_parte3 = nlp(text_parte3)
        
        # Itera sobre cada palavra no documento processado
        for token in doc_parte1:
            # Verifica se a palavra é uma localidade
            if token.ent_type_ == "LOC":
                categorias_aux.append("localidade")
                texto_aux.append(token)
            else:
                categorias_aux.append("não_localizada")
                texto_aux.append(token)
                
        # Itera sobre cada palavra no documento processado
        for token in doc_parte2:
            # Verifica se a palavra é uma localidade
            if token.ent_type_ == "LOC":
                categorias_aux.append("localidade")
                texto_aux.append(token)
            else:
                categorias_aux.append("não_localizada")
                texto_aux.append(token)
                
        # Itera sobre cada palavra no documento processado
        for token in doc_parte3:
            # Verifica se a palavra é uma localidade
            if token.ent_type_ == "LOC":
                categorias_aux.append("localidade")
                texto_aux.append(token)
            else:
                categorias_aux.append("não_localizada")
                texto_aux.append(token)
        
    else:
        # DOC
        doc = nlp(df['Texto'][i])
        # Itera sobre cada palavra no documento processado
        for token in doc:
            # Verifica se a palavra é uma localidade
            if token.ent_type_ == "LOC":
                categorias_aux.append("localidade")
                texto_aux.append(token)
            else:
                categorias_aux.append("não_localizada")
                texto_aux.append(token)
            
    texto.append(texto_aux)
    categorias.append(categorias_aux)
```

```python
frases_lema_sem_stops = []
for i in range(len(lista_texto)):
    a = [elemento for elemento in lista_texto[i] if elemento not in stopps]
    frases_lema_sem_stops.append(a)
```
